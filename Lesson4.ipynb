# Задание 2

```python
t = np.linspace(0,10,51)
f = np.cos(t)
plt.plot(t,f, color = '#77ff99') # такой цвет кажется мне куда приятнее, чем #00FF00
plt.title('График f(t)', fontsize = 14)
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5,9.5,-2.5,2.5])
plt.show()
```

# Задание 3

```python
x = np.linspace(-3,3,51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x + 1.5
y4 = np.sin(x)

fig,ax = plt.subplots(nrows = 2, ncols = 2)
ax_row1, ax_row2 = ax
ax11, ax12 = ax_row1
ax21, ax22 = ax_row2


ax11.plot(x,y1)
ax11.set_title('График y1(x)', fontsize = 14)
ax11.set_xlim([-5,5])

ax12.set_title('График y2(x)', fontsize = 14)
ax12.plot(x,y2)

ax21.set_title('График y3(x)', fontsize = 14)
ax21.plot(x,y3)

ax22.set_title('График y4(x)', fontsize = 14)
ax22.plot(x,y4)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
```
